```python
def StagePredictor(top1Path, picklePath, allFeaturesDF):
  top1DF = pd.read_csv(top1Path)
  top1Record = top1DF[top1DF["Order"] == 1]
  top1Cls = top1Record["Classifier"].values[0]
  top1Scaler = top1Record["Scaler"].values[0]
  top1 = (top1Cls, top1Scaler)

  data = LoadPickle(picklePath)
  history = data[0]

  finalPredictions = []
  historyPredictions = []
  for key in history.keys():
    metaColsDrop = history[key]["Meta_Columns_Dropped"]
    dropFeatures = history[key]["Drop_Features"]
    keepFeatures = history[key]["Keep_Features"]
    encoderLabels = history[key]["Encoder_Labels"]
    classification = history[key]["Classification"]
    modelsObj = classification[top1]["Models_OBJ"]
    scalerObj = classification[top1]["Scaler_OBJ"]
    featuresRecord = allFeaturesDF[
      (allFeaturesDF["Quantile"] == key[0]) & (allFeaturesDF["Position"] == key[1])][keepFeatures
    ].values
    featuresRecordScaled = scalerObj.transform(featuresRecord)
    preds = []
    for i in range(len(modelsObj)):
      preds.append(modelsObj[0].predict(featuresRecordScaled)[0])
    majorPred = collections.Counter(preds).most_common(1)[0][0]
    finalPredictions.append(majorPred)
    historyPredictions.append({
      "Quantile"  : key[0],
      "Position"  : key[1],
      "Prediction": majorPred,
      "Label"     : encoderLabels[majorPred],
    })
  finalPrediction = collections.Counter(finalPredictions).most_common(1)[0][0]
  finalPredictionLabel = encoderLabels[finalPrediction]

  del top1DF, top1Record, top1Cls, top1Scaler, top1, data, history, finalPredictions

  return historyPredictions, finalPrediction, finalPredictionLabel


def InferencePhase(selectedRun, imgPath, maskPath):
  configsPath = os.path.join(selectedRun, r"Configurations.pickle")

  with open(configsPath, "rb") as f:
    configurations = pickle.load(f)

  featuresConfigsPath = os.path.join(featuresPath, configurations["Features File"], "Configurations.pickle")

  with open(featuresConfigsPath, "rb") as f:
    featuresConfigurations = pickle.load(f)
  distancesStr = featuresConfigurations["Distances"]
  distancesList = distancesStr.split(",")
  distancesList = [int(d) for d in distancesList]

  image = cv2.imread(imgPath, cv2.IMREAD_COLOR)
  mask = cv2.imread(maskPath, cv2.IMREAD_GRAYSCALE)

  mask = Gray2Binary(mask)
  maskCopy = mask.copy()

  # Find the contours.
  cnts = cv2.findContours(maskCopy, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
  cnts = imutils.grab_contours(cnts)
  cnts = sorted(cnts, key=cv2.contourArea)

  if (len(cnts) < 3):
    print("Less than 3 contours.")
    return None

  try:
    allFeatures = []
    for j, c in enumerate(cnts):
      folder = "Bottom"
      if (j == 1):
        folder = "Middle"
      elif (j == 2):
        folder = "Top"

      featuresDict = CalculateContourFeatures(c)
      subMaskCopy = np.zeros_like(mask)
      cv2.fillPoly(subMaskCopy, [c], color=(255, 255, 255))
      distanceMap = CalculateDistanceMap(subMaskCopy)
      quantiles = np.quantile(
        sorted(set(distanceMap.flatten())),
        np.array(distancesList) / 100.0,
      )

      for q, quantile in enumerate(quantiles):
        subFeatures = {}

        temp = subMaskCopy.copy()
        temp[distanceMap > quantile] = 0
        distance = distancesList[q]
        roi = cv2.bitwise_and(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), temp)

        data = [(roi, temp)]

        for z, (a, b) in enumerate(data):
          roiNoPad = RemovePadding(a).astype(np.uint8)
          tempNoPad = RemovePadding(b).astype(np.uint8)

          firstK, firstV = RadiomicsFeaturesHelper().CalculateFirstOrderFeatures(
            roiNoPad,
            isSingleObject=True,
            returnSqueezedRowDict=True,
            returnKeyValuesLists=True,
            discardKeys=["EmpiricalCumulativeDistributionFunction"],
          )

          glrlm = RadiomicsFeaturesHelper().CalculateMultiGLRLM(
            roiNoPad,
            angles=[0, 45, 90, 135],  # 4 Angles.
            includeZero=True,
            normalize=True,
            applyAverage=True,
            grayLevels=(0, 255),
          )

          glcm = RadiomicsFeaturesHelper().CalculateMultiGLCM(
            roiNoPad,
            angles=[0, 45, 90, 135],  # 4 Angles.
            distances=[1, 3, 5],  # 3 Distances.
            normalize=True,
            applyAverage=True,
            grayLevels=(0, 255),
          )

          glrlmFeatures = RadiomicsFeaturesHelper().CalculateGLRLMAllFeatures(glrlm)
          glcmFeatures = RadiomicsFeaturesHelper().CalculateGLCMAllFeatures(glcm)

          for ii, kk in enumerate(firstK):
            subFeatures[kk] = firstV[ii]
          for kk in glcmFeatures.keys():
            subFeatures[kk] = glcmFeatures[kk]
          for kk in glrlmFeatures.keys():
            subFeatures[kk] = glrlmFeatures[kk]
          for kk in featuresDict.keys():
            subFeatures[kk] = featuresDict[kk]

        subFeatures["Quantile"] = distance
        subFeatures["Position"] = folder
        allFeatures.append(subFeatures)

    allFeaturesDF = pd.DataFrame(allFeatures)

    fsTop1Path = os.path.join(selectedRun, r"First_Stage_Top1_Individual_Metrics.csv")
    fsPicklePath = os.path.join(selectedRun, r"Meta Results\First_Stage_History.pickle")

    secUpTop1Path = os.path.join(selectedRun, r"Second_Stage_Up_Top1_Individual_Metrics.csv")
    secUpPicklePath = os.path.join(selectedRun, r"Meta Results\Second_Stage_Up_History.pickle")

    secDownTop1Path = os.path.join(selectedRun, r"Second_Stage_Down_Top1_Individual_Metrics.csv")
    secDownPicklePath = os.path.join(selectedRun, r"Meta Results\Second_Stage_Down_History.pickle")

    print("First Stage:")
    fsHistoryPredictions, fsFinalPrediction, fsFinalPredictionLabel = StagePredictor(
      fsTop1Path,
      fsPicklePath,
      allFeaturesDF
    )
    for el in fsHistoryPredictions:
      a, b, c, d = el.values()
      print(f"Quantile: {a}, Position: {b}, Prediction: {c}, Label: {d}")
    print("Final Prediction:", fsFinalPrediction)
    print("Final Prediction Label:", fsFinalPredictionLabel)

    print("Second Stage:")
    if (fsFinalPrediction == 0):
      historyPredictions, finalPrediction, finalPredictionLabel = StagePredictor(
        secUpTop1Path,
        secUpPicklePath,
        allFeaturesDF
      )
    elif (fsFinalPrediction == 1):
      historyPredictions, finalPrediction, finalPredictionLabel = StagePredictor(
        secDownTop1Path,
        secDownPicklePath,
        allFeaturesDF
      )
    else:
      raise Exception("Invalid First Stage Prediction.")
    for el in historyPredictions:
      a, b, c, d = el.values()
      print(f"Quantile: {a}, Position: {b}, Prediction: {c}, Label: {d}")
    print("Final Prediction:", finalPrediction)
    print("Final Prediction Label:", finalPredictionLabel)

    resultsDict = {
      "Features"                           : allFeaturesDF,
      "First_Stage_History"                : fsHistoryPredictions,
      "First_Stage_Final_Prediction"       : fsFinalPrediction,
      "First_Stage_Final_Prediction_Label" : fsFinalPredictionLabel,
      "Second_Stage_History"               : historyPredictions,
      "Second_Stage_Final_Prediction"      : finalPrediction,
      "Second_Stage_Final_Prediction_Label": finalPredictionLabel,
      "Contours"                           : cnts,
      "Image"                              : image,
      "Mask"                               : mask,
      "Distances"                          : distancesList,
    }

    plt.figure()
    distancesList = list(sorted(resultsDict["Distances"]))[::-1]
    positions = ["Bottom", "Middle", "Top"][::-1]

    counter = 0
    overAllMask = np.zeros_like(resultsDict["Mask"])
    overAllMaskRGB = cv2.cvtColor(overAllMask, cv2.COLOR_GRAY2RGB)
    for cntNumber in range(len(positions)):
      subMask = np.zeros_like(resultsDict["Mask"])
      cv2.fillPoly(subMask, [resultsDict["Contours"][cntNumber]], color=(255, 255, 255))
      distanceMap = CalculateDistanceMap(subMask)
      quantiles = np.quantile(
        sorted(set(distanceMap.flatten())),
        np.array(distancesList) / 100.0,
      )
      for quanNumber in range(len(distancesList)):
        tempMask = subMask.copy()
        tempMask[distanceMap > quantiles[quanNumber]] = 0
        label = [
          item['Label']
          for item in resultsDict["Second_Stage_History"]
          if ((item['Quantile'] == distancesList[quanNumber]) and (item['Position'] == positions[cntNumber]))
        ][0]
        rndColor = np.random.randint(100, 255, 3)
        overAllMaskRGB[tempMask == 255] = rndColor
        plt.text(
          50,
          75 + 12 * (7 * counter + 1),
          f"{positions[cntNumber]} - {distancesList[quanNumber]}% => {label}",
          color=rndColor / 255,
          fontsize=8,
          fontweight="bold",
        )
        counter += 1

    plt.text(
      50,
      75 + 12 * (7 * counter + 1),
      f"Final Prediction: {resultsDict['Second_Stage_Final_Prediction_Label']}",
      color="white",
      fontsize=8,
      fontweight="bold",
    )
    plt.imshow(cv2.cvtColor(overAllMaskRGB, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.tight_layout()
    # Store the plot without displaying it.
    resultsDict["Plot"] = plt

    return resultsDict
  except Exception as e:
    print(e)
    return None
```
